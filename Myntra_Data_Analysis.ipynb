**MYNTRA DATA ANALYSIS**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel("Myntra dataset 11.xlsx")
df
```

```python
customers = pd.read_excel("Myntra dataset 11.xlsx" , sheet_name = "dim_customers")
customers
```

```python
orders = pd.read_excel("Myntra dataset 11.xlsx", sheet_name = "fact_orders")
orders
```

```python
products = df
products
```

```python
products.info()
```

```python
#Check for duplicates

products.duplicated().sum()
customers.duplicated().sum()
orders.duplicated().sum()
```

```python
customers.isnull().sum()
```

```python
customers.dropna(axis = 0, inplace=True)
customers.isnull().sum()
```

```python
orders.describe()
```

```python
orders["Month"] = orders["Date"].dt.strftime("%B")
orders
```

```python
orders["Total Price"] = orders["Original Price"] - (orders["Original Price"] * orders["Discount%"])
orders
```

```python
gb = orders.groupby("Month").agg({"Total Price":"sum"})
gb
```

```python
#plt.figure(figsize = (12,4))
sns.barplot(x = gb.index, y = gb["Total Price"], order = ["January", "February", "March",'April',"May","June","July","August","September","October","November", "December"],hue = gb.index, palette = "GnBu")
plt.xticks(rotation = 45)
plt.show()
```

```python
# Convert the 'Size' column to numeric, handling non-numeric values
products['Size'] = pd.to_numeric(products['Size'], errors='coerce')

# Group by 'Color' and calculate the sum of 'Size' (now numeric)
gb1 = products.groupby("Color").agg({"Size": "sum"})
gb1
```

```python
sns.barplot(x=gb1.index, y=gb1["Size"], palette="viridis")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=np.number)
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Category'], bins=30, kde=True)
plt.title('Product Name')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
if 'Sub-Category' in df.columns:
    sns.countplot(data=df, x='Brand Name', order=df['Sub-Category'].value_counts().index)
else:

    print("Column 'Sub-Category' not found in DataFrame. Using default order.")
    sns.countplot(data=df, x='Brand Name')

plt.title('Count of Products by Category')
plt.xticks(rotation=90)
plt.show()
```

```python
total_original_price = orders['Original Price'].sum()
total_discount = orders['Total Price'].sum()

labels = ['Original Price', 'Discount']
sizes = [total_original_price, total_original_price-total_discount]
colors = ['skyblue', 'lightcoral']
explode = (0.1, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.title('Original Price vs. Discount')

plt.show()
```

```python
sns.violinplot(x='State', y='City', data=customers , palette='Set2')
plt.xlabel('State')
plt.ylabel('City')
plt.xticks(rotation=45)
plt.title('Distribution of City by State')
plt.show()

plt.show()
```

```python
print(products.columns)
if 'Rating' not in products.columns:
    sns.boxplot(x="Color", y="Ratings", data=products, palette="inferno")

else:
    sns.boxplot(x="Color", y="Ratings", data=products, palette="inferno")

plt.xticks(rotation=90)
plt.show()
```

```python
sns.histplot(products['Ratings'], bins=10, kde=True)
plt.title('Distribution of Ratings')
plt.xlabel('Ratings')
plt.ylabel('Frequency')
plt.show()

sns.scatterplot(x='Size', y='Ratings', data=products, hue='Color')
plt.title('Relationship between Size and Ratings')
plt.show()
```
